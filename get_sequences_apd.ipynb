```python
url = r'http://aps.unmc.edu/AP/database/query_output.php?ID='
rows =[]
columns = []
all_tables = []
for i in range(1, 3248):
    new_url = url+str(i).zfill(5)
    print(new_url)
    tables = pd.read_html(new_url) # Returns list of all tables on page
    desired_table = tables[2]
    all_tables.append(desired_table)
    columns = desired_table[1]
    rows.append(list(desired_table[2]))
   
#df = df.append(pd.Series(rows, index=df.columns[:len(rows)]), ignore_index=True)
#df = pd.concat(rows, ignore_index=True)   
#df = pd.DataFrame(rows, columns=columns) 
```

```python
#Going through all tables, building dataframes and treating null columns.

ready_df = [] # It will store ready dataframes and concatenate them later.
for table in all_tables:
    new_table = table.filter(items=[1,2]) #table with coluns=1 and rows=2
    new_df = pd.DataFrame([list(new_table[2])], columns=new_table[1]) #make new df.
    new_df = new_df.dropna(axis=1, how='any') #removing NaN columns.
    new_df = new_df.filter(items=['Sequence:','Length:', 'Net Charge:','Net charge:','Hydrophobic residue%:','Boman Index:','3D Structure:','Activity:'])
    ready_df.append(new_df)
```

```python
final_df = pd.concat(ready_df, axis=0, join='outer', ignore_index=True, keys=None,
          levels=None, names=None, verify_integrity=False, copy=True, sort=True)
```
