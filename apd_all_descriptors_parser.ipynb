```python
import pandas as pd
```

```python
#apd_sequences = pd.read_csv("apd_sequences_all_descriptors.csv")
apd_sequences = pd.read_csv("helix_sequences_all_descriptors.csv")
```

```python
apd_sequences
```

```python
#All columns has "'b" and "'" in all values.
for column in apd_sequences.columns:
    print(column)
    apd_sequences[column] = apd_sequences[column].apply(lambda x: x.replace("b'","").replace("'", "")) 
```

```python
apd_sequences
```

```python
apd_sequences.head()
```

```python
apd_sequences.columns= ['Sequence__','Length', 'MW', 'Charge', 'ChargeDensity', 'pI', 'InstabilityInd', 'Aromaticity', 'AliphaticInd', 'BomanInd', 'HydrophRatio']
```

```python
apd_sequences
```

```python
#All columns are floating.
float_columns = ['MW', 'Charge', 'ChargeDensity', 'pI', 'InstabilityInd', 'Aromaticity', 'AliphaticInd', 'BomanInd', 'HydrophRatio']
for column in float_columns:
    apd_sequences[column] = apd_sequences[column].apply(lambda x: float(x))
```

```python
apd_sequences
```

```python
apd_sequences = apd_sequences.drop(columns=['Length'])
```

```python
apd_sequences
```

```python
#apd_sequences.to_csv('apd_sequences_descriptor_final.csv')
apd_sequences.to_csv('helix_sequences_descriptor_final.csv')
```

```python
#df = pd.read_csv('apd_df.csv')
df = pd.read_csv('helix_sequences.csv')
```

```python
final_df = pd.concat([df, apd_sequences], axis=1, join='inner')
```

```python
final_df.sample(60)
```

```python
final_df = final_df.drop(columns=['Unnamed: 0','Sequence__', 'Charge','Boman_Index'])
```
